```python
import pandas as pd
```

## Lithium

```python
# Load lithium data
df_lithium = pd.read_csv('data/ALL_lithium_prices_cubic_spline.csv')
df_lithium = df_lithium.drop(columns=['LISAME'])
df_lithium['Date'] = pd.to_datetime(df_lithium['Date'])
# df_lithium = df_lithium[df_lithium['Date'] >= '2021-08-01']
df_lithium.head()
```

```python
df_lithium_returns = df_lithium.drop(columns=['Date']).pct_change()

# Plot correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix_lithium = df_lithium_returns.corr()

start_date = df_lithium['Date'].min().strftime('%Y-%m-%d')
end_date = df_lithium['Date'].max().strftime('%Y-%m-%d')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_lithium, annot=True, cmap='RdYlGn', center=0, fmt=".2f")
plt.title('Correlation Heatmap of Lithium Price Returns\nfrom {} to {}'.format(start_date, end_date))
plt.show()
```

## Cobalt

```python
# Load cobalt data
df_cobalt = pd.read_csv('data/ALL_cobalt_prices_cubic_spline.csv')
df_cobalt['Date'] = pd.to_datetime(df_cobalt['Date'])
# only dates until end of 2024
df_cobalt = df_cobalt[df_cobalt['Date'] <= '2024-12-31']
df_cobalt = df_cobalt[df_cobalt['Date'] >= '2017-01-03']
df_cobalt.head()
```
