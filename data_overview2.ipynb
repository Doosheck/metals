```python
import pandas as pd
```

## Lithium

```python
# Load lithium data
df_lithium = pd.read_csv('data/ALL_lithium_prices_cubic_spline.csv')
df_lithium = df_lithium.drop(columns=['LISAME'])
df_lithium['Date'] = pd.to_datetime(df_lithium['Date'])
df_lithium = df_lithium[df_lithium['Date'] >= '2021-08-01']
df_lithium.head()
```

```python
# Plot correlation heatmap
import seaborn as sns
import matplotlib.pyplot as plt
corr_matrix = df_lithium.drop(columns=['Date']).corr()

start_date = df_lithium['Date'].min().strftime('%Y-%m-%d')
end_date = df_lithium['Date'].max().strftime('%Y-%m-%d')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', center=0, fmt=".2f")
plt.title('Correlation Heatmap of Lithium Prices (cubic spline interpolated)\nfrom {} to {}'.format(start_date, end_date))
plt.show()
```
