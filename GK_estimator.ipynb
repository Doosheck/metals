```python
cobalt_hloc = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/cobalt_hloc_2017_2024.csv', sep=';', header=0)
cobalt_hloc = cobalt_hloc.drop(0)
cobalt_hloc.head()
```

```python
# Columns are PX_LAST, PX_HIGH, PX_LOW, PX_OPEN
# Based on those compute the garman-klass volatility
# Replace commas with dots and convert to float
cobalt_hloc['PX_LAST'] = cobalt_hloc['PX_LAST'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_HIGH'] = cobalt_hloc['PX_HIGH'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_LOW'] = cobalt_hloc['PX_LOW'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_OPEN'] = cobalt_hloc['PX_OPEN'].astype(str).str.replace(',', '.').astype(float)

cobalt_hloc['log_ret'] = cobalt_hloc['PX_LAST'].pct_change()
cobalt_hloc['log_ret'] = cobalt_hloc['log_ret'].fillna(0)
cobalt_hloc['log_ret'] = cobalt_hloc['log_ret'].astype(float)

cobalt_hloc['volatility'] = cobalt_hloc['log_ret'].rolling(window=252).std() * (252**0.5)
cobalt_hloc['volatility'] = cobalt_hloc['volatility'].fillna(0)
cobalt_hloc['volatility'] = cobalt_hloc['volatility'].astype(float)

cobalt_hloc.head()
```

```python
copper_hloc = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/copper_hloc_2017_2024.csv', sep=';', header=0)
copper_hloc = copper_hloc.drop(0)
copper_hloc['PX_LAST'] = copper_hloc['PX_LAST'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_HIGH'] = copper_hloc['PX_HIGH'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_LOW'] = copper_hloc['PX_LOW'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_OPEN'] = copper_hloc['PX_OPEN'].astype(str).str.replace(',', '.').astype(float)

copper_hloc['log_ret'] = copper_hloc['PX_LAST'].pct_change()
copper_hloc['log_ret'] = copper_hloc['log_ret'].fillna(0)
copper_hloc['log_ret'] = copper_hloc['log_ret'].astype(float)

copper_hloc['volatility'] = copper_hloc['log_ret'].rolling(window=252).std() * (252**0.5)
copper_hloc['volatility'] = copper_hloc['volatility'].fillna(0)
copper_hloc['volatility'] = copper_hloc['volatility'].astype(float)

copper_hloc.head()
```

```python
nickel_hloc = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/nickel_HLOC_2017_2024.csv', sep=';', header=0)
nickel_hloc = nickel_hloc.drop(0)
nickel_hloc['PX_LAST'] = nickel_hloc['PX_LAST'].astype(str).str.replace(',', '.').astype(float)
nickel_hloc['PX_HIGH'] = nickel_hloc['PX_HIGH'].astype(str).str.replace(',', '.').astype(float)
nickel_hloc['PX_LOW'] = nickel_hloc['PX_LOW'].astype(str).str.replace(',', '.').astype(float)
nickel_hloc['PX_OPEN'] = nickel_hloc['PX_OPEN'].astype(str).str.replace(',', '.').astype(float)
nickel_hloc['log_ret'] = nickel_hloc['PX_LAST'].pct_change()
nickel_hloc['log_ret'] = nickel_hloc['log_ret'].fillna(0)
nickel_hloc['log_ret'] = nickel_hloc['log_ret'].astype(float)
nickel_hloc['volatility'] = nickel_hloc['log_ret'].rolling(window=252).std() * (252**0.5)
nickel_hloc['volatility'] = nickel_hloc['volatility'].fillna(0)
nickel_hloc['volatility'] = nickel_hloc['volatility'].astype(float)
nickel_hloc.head()
```

```python
# Compute the Garman-Klass estimator for Copper

def calc_gk_estimator(data):
    nobs = len(data)
    estimator = 0

    for i in range(1, nobs):
        estimator += 0.5 * (np.log(data['PX_HIGH'][i] / data['PX_LOW'][i])**2) - (2 * np.log(2) - 1) * np.log(data['PX_LAST'][i] / data['PX_OPEN'][i])**2
    if estimator < 0:
        raise ValueError("Estimator value is negative, cannot compute square root.")
    estimator = np.sqrt(estimator)
    return estimator/np.sqrt(nobs)
```

```python
copper_gk = calc_gk_estimator(copper_hloc)
cobalt_gk = calc_gk_estimator(cobalt_hloc)
nickel_gk = calc_gk_estimator(nickel_hloc)

print(f"Copper GK estimator: {copper_gk}")
print(f"Cobalt GK estimator: {cobalt_gk}")
print(f"Nickel GK estimator: {nickel_gk}")
```

```python
def calc_parkinson_estimator(data):
    """
    Compute the Parkinson volatility estimator.

    Parameters
    ----------
    data : pandas.DataFrame
        Must contain columns 'PX_HIGH' and 'PX_LOW' for each period.

    Returns
    -------
    float
        Annualized Parkinson volatility estimate (per period sqrt(1) basis).

    Raises
    ------
    ValueError
        If the computed variance is negative (numerical issues).
    """
    # Number of observations
    nobs = len(data)
    if nobs == 0:
        raise ValueError("Input data must contain at least one observation.")
    
    # Sum of squared log high/low ratios
    sum_sq = np.sum(
        np.log(data['PX_HIGH'] / data['PX_LOW']) ** 2
    )
    
    # Parkinson variance estimator
    variance = sum_sq / (4 * nobs * np.log(2))
    if variance < 0:
        raise ValueError("Negative variance encountered; check input data.")
    
    # Volatility is the square root of variance
    return np.sqrt(variance)
```

```python
# Compute the Parkinson estimator for Copper
copper_parkinson = calc_parkinson_estimator(copper_hloc)
cobalt_parkinson = calc_parkinson_estimator(cobalt_hloc)
nickel_parkinson = calc_parkinson_estimator(nickel_hloc)
print(f"Copper Parkinson estimator: {copper_parkinson}")
print(f"Cobalt Parkinson estimator: {cobalt_parkinson}")
print(f"Nickel Parkinson estimator: {nickel_parkinson}")
```
