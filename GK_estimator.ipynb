```python
# Data overview

import pandas as pd
import numpy as np
import openpyxl

# Load data
cobalt = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/cobalt_ask_bid_2017_2024.csv', sep=';')
cobalt.head()
```

```python
cobalt_hloc = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/cobalt_hloc_2017_2024.csv', sep=';', header=0)
cobalt_hloc = cobalt_hloc.drop(0)
cobalt_hloc.head()
```

```python
# Columns are PX_LAST, PX_HIGH, PX_LOW, PX_OPEN
# Based on those compute the garman-klass volatility
# Replace commas with dots and convert to float
cobalt_hloc['PX_LAST'] = cobalt_hloc['PX_LAST'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_HIGH'] = cobalt_hloc['PX_HIGH'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_LOW'] = cobalt_hloc['PX_LOW'].astype(str).str.replace(',', '.').astype(float)
cobalt_hloc['PX_OPEN'] = cobalt_hloc['PX_OPEN'].astype(str).str.replace(',', '.').astype(float)

cobalt_hloc['log_ret'] = cobalt_hloc['PX_LAST'].pct_change()
cobalt_hloc['log_ret'] = cobalt_hloc['log_ret'].fillna(0)
cobalt_hloc['log_ret'] = cobalt_hloc['log_ret'].astype(float)

cobalt_hloc['volatility'] = cobalt_hloc['log_ret'].rolling(window=252).std() * (252**0.5)
cobalt_hloc['volatility'] = cobalt_hloc['volatility'].fillna(0)
cobalt_hloc['volatility'] = cobalt_hloc['volatility'].astype(float)

cobalt_hloc.head()
```

```python
copper_hloc = pd.read_csv('/Users/michal/Documents/Code/metals/bloomberg_data/copper_hloc_2017_2024.csv', sep=';', header=0)
copper_hloc = copper_hloc.drop(0)
copper_hloc['PX_LAST'] = copper_hloc['PX_LAST'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_HIGH'] = copper_hloc['PX_HIGH'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_LOW'] = copper_hloc['PX_LOW'].astype(str).str.replace(',', '.').astype(float)
copper_hloc['PX_OPEN'] = copper_hloc['PX_OPEN'].astype(str).str.replace(',', '.').astype(float)

copper_hloc['log_ret'] = copper_hloc['PX_LAST'].pct_change()
copper_hloc['log_ret'] = copper_hloc['log_ret'].fillna(0)
copper_hloc['log_ret'] = copper_hloc['log_ret'].astype(float)

copper_hloc['volatility'] = copper_hloc['log_ret'].rolling(window=252).std() * (252**0.5)
copper_hloc['volatility'] = copper_hloc['volatility'].fillna(0)
copper_hloc['volatility'] = copper_hloc['volatility'].astype(float)

copper_hloc.head()
```

```python
# Compute the Garman-Klass estimator for Copper

def calc_gk_estimator(data):
    nobs = len(data)
    estimator = 0

    for i in range(1, nobs):
        estimator += 0.5 * (np.log(data['PX_HIGH'][i] / data['PX_LOW'][i])**2 - (2 * np.log(2) - 1) * np.log(data['PX_LAST'][i] / data['PX_OPEN'][i])**2)

    return estimator
```

```python
copper_gk = calc_gk_estimator(copper_hloc)
cobalt_gk = calc_gk_estimator(cobalt_hloc)
print(f"Copper GK estimator: {copper_gk}")
print(f"Cobalt GK estimator: {cobalt_gk}")
```
