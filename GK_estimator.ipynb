```python
# Compute the Garman-Klass estimator for Copper

def calc_gk_estimator(data):
    nobs = len(data)
    estimator = 0

    for i in range(1, nobs):
        estimator += 0.5 * (np.log(data['PX_HIGH'][i] / data['PX_LOW'][i])**2) - (2 * np.log(2) - 1) * np.log(data['PX_LAST'][i] / data['PX_OPEN'][i])**2
    if estimator < 0:
        raise ValueError("Estimator value is negative, cannot compute square root.")
    estimator = np.sqrt(estimator)
    return estimator/np.sqrt(nobs)
```

```python
copper_gk = calc_gk_estimator(copper_hloc)
cobalt_gk = calc_gk_estimator(cobalt_hloc)
nickel_gk = calc_gk_estimator(nickel_hloc)

print(f"Copper GK estimator: {copper_gk}")
print(f"Cobalt GK estimator: {cobalt_gk}")
print(f"Nickel GK estimator: {nickel_gk}")
```

```python
def calc_parkinson_estimator(data):
    """
    Compute the Parkinson volatility estimator.

    Parameters
    ----------
    data : pandas.DataFrame
        Must contain columns 'PX_HIGH' and 'PX_LOW' for each period.

    Returns
    -------
    float
        Annualized Parkinson volatility estimate (per period sqrt(1) basis).

    Raises
    ------
    ValueError
        If the computed variance is negative (numerical issues).
    """
    # Number of observations
    nobs = len(data)
    if nobs == 0:
        raise ValueError("Input data must contain at least one observation.")
    
    # Sum of squared log high/low ratios
    sum_sq = np.sum(
        np.log(data['PX_HIGH'] / data['PX_LOW']) ** 2
    )
    
    # Parkinson variance estimator
    variance = sum_sq / (4 * nobs * np.log(2))
    if variance < 0:
        raise ValueError("Negative variance encountered; check input data.")
    
    # Volatility is the square root of variance
    return np.sqrt(variance)
```

```python
# Compute the Parkinson estimator for Copper
copper_parkinson = calc_parkinson_estimator(copper_hloc)
cobalt_parkinson = calc_parkinson_estimator(cobalt_hloc)
nickel_parkinson = calc_parkinson_estimator(nickel_hloc)
print(f"Copper Parkinson estimator: {copper_parkinson}")
print(f"Cobalt Parkinson estimator: {cobalt_parkinson}")
print(f"Nickel Parkinson estimator: {nickel_parkinson}")
```

```python
def compute_liquidity_metrics(df):
    """
    Compute liquidity metrics from a DataFrame with bid/ask data.

    Parameters
    ----------
    df : pd.DataFrame
        Must contain columns 'PX_ASK' and 'PX_BID'.

    Returns
    -------
    pd.DataFrame
        Original data plus:
        - 'spread': absolute ask-bid spread
        - 'mid': midpoint price
        - 'pct_spread_ask': spread divided by ask price
        - 'pct_spread_mid': spread divided by midpoint
    """
    # Absolute spread
    df['spread'] = df['PX_ASK'] - df['PX_BID']
    # Midpoint price
    df['mid'] = (df['PX_ASK'] + df['PX_BID']) / 2
    # Percentage spread relative to ask
    df['pct_spread_ask'] = df['spread'] / df['PX_ASK']
    # Percentage spread relative to midpoint
    df['pct_spread_mid'] = df['spread'] / df['mid']
    
    return df
```

```python
cobalt_liquidity = compute_liquidity_metrics(cobalt_bid_ask)
cobalt_liquidity
```
